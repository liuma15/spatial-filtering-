# Spatial Filtering

**We use different types of filtering such as local averaging, Gaussian filter, Laplacian filtering and Median filetring to reduce noise and blur an image.***

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, data, filters, morphology
from skimage.filters import *
from skimage.morphology import square
from skimage import util
```

## Box filter or mean filter

```python
plt.subplot(1,2,1)
img = color.rgb2gray(io.imread('http://astropetros.eu/wp/wp-content/uploads/2017/06/Moon_213655_g4_ap558_proc.jpg'))
io.imshow(img)
plt.axis('off')

plt.subplot(1,2,2)
img_mean = rank.mean(img, square(11)) # (image, selem): selem is the neighborhood expressed as an ndarray of 1s and 0s. Square: generates a flat, square-shaped structuring element.
io.imshow(img_mean)
plt.axis('off')
```

## TODO: After running the code in the above cell, write code below to show the input and output images using mean filters with filter size 5x5, 15x15, 25x25, respectively.

```python
## TODO: write code here
plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
img_mean = rank.mean(img, square(5)) 
io.imshow(img_mean)
plt.axis('off')

plt.subplot(1,3,2)
img_mean = rank.mean(img, square(15)) 
io.imshow(img_mean)
plt.axis('off')

plt.subplot(1,3,3)
img_mean = rank.mean(img, square(25)) 
io.imshow(img_mean)
plt.axis('off')
plt.show()
```

**The quality of the image reduces as we use a larger neighborhood for the local averaging.**

## Gaussian filter


```python
plt.subplot(1,2,1)
img = color.rgb2gray(io.imread('http://astropetros.eu/wp/wp-content/uploads/2017/06/Moon_213655_g4_ap558_proc.jpg'))
io.imshow(img)
plt.axis('off')

plt.subplot(1,2,2)
img_mean = gaussian(img, 1) 
io.imshow(img_mean)
plt.axis('off')
```

## Laplacian filter

```python
# Find the edges of an image using the Laplace operator
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
img = data.camera()
io.imshow(img)
plt.axis('off')

plt.subplot(1,2,2)
laplace_edges =laplace(img) # (image, ksize=3, mask=None)
io.imshow(laplace_edges, cmap='gray')
plt.axis('off')
```

## TODO: After running the code in the above cell, write code below to show the input and output images using Laplacian filters with filter size=5, filter size=35, respectively.

```python
## TODO: write code here
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
laplace_edges =laplace(img, ksize=5)
io.imshow(laplace_edges, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
laplace_edges =laplace(img, ksize=35) 
io.imshow(laplace_edges, cmap='gray')
plt.axis('off')
plt.show()
```

## Median filter

```python
img = data.camera()
plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
io.imshow(img)
plt.axis('off')

salt_noise_img = util.random_noise(img,'salt')
plt.subplot(1,3,2)
io.imshow(salt_noise_img)
plt.axis('off')

denoised_image = median(salt_noise_img, square(3))
plt.subplot(1,3,3)
io.imshow(denoised_image)
plt.axis('off')
```
